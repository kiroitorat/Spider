##使用py自己构造请求
-- 需要一个库：requests

```python
import requests
```

#请求包括：
请求行
请求头
请求体

```python
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}
```

```python
url = "https://www.xinpianchang.com/discover/article-9999-0-all-all-0-0-score?utm_source=xpcWeb&utm_medium=navigatorCate&index=3"
```

```python
#第一个参数是方法，第二个参数是地址
response = requests.request("get", url, headers = headers)
```

网络请求的状态码：
-2XX开头：成功系列

-3XX：重定向

-4XX:客户端错误

-5XX：服务器内部错误

```python
response.text
```
